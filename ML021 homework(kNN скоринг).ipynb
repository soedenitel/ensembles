Примените метод ближайших соседей (kNN) для классификации скоринга, используйте k=100. Используйте биометрические данные, все столбцы Insurance_History, Family_Hist, Medical_History и InsurеdInfo. Заполните отсутствующие значения -1.

Проведите предсказание и проверьте качество через каппа-метрику.

Данные: video.ittensive.com/machine-learning/prudential/train.csv.gz

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import cohen_kappa_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
import re
from etl_utils import reduce_mem_usage, show_inf_and_na, inf_and_na_columns
pd.set_option('display.max_columns', 200)
```

Загрузите данные.

```python
data_df = reduce_mem_usage(pd.read_csv("https://video.ittensive.com/machine-learning/prudential/train.csv.gz"))
data_df.head()
```

Используйте биометрические данные, все столбцы Insurance_History, Family_Hist, Medical_History и InsurеdInfo

Заполните отсутствующие значения -1.

```python
feature_regsearcher = r'Insurance_History.*|Family_Hist.*|Medical_History.*|InsuredInfo.*|Wt|Ht|Ins_Age|BMI'
x_features = [feature for feature in data_df.columns if re.match(feature_regsearcher, feature) != None]
data_df[x_features] = data_df[x_features].fillna(-1)
show_inf_and_na(data_df[x_features])
```

Разделите выборку на обучающую/проверочную в соотношении 80/20.

```python
train_df, test_df = train_test_split(data_df, test_size=.2)
```

Примените метод ближайших соседей (kNN) для классификации скоринга, используйте k=100.

```python
knn = KNeighborsClassifier(n_neighbors=100, n_jobs=-1)
knn.fit(train_df[x_features], train_df['Response'])
```

Проведите предсказание и проверьте качество через каппа-метрику.

```python
predictions = knn.predict(test_df[x_features])
kappa = cohen_kappa_score(test_df['Response'], predictions, weights='quadratic')
print('knn 100 каппа:', kappa)
```

Какое значение каппа метрики получившейся модели с точностью до сотых ?

```python
round(kappa, 2)
```
